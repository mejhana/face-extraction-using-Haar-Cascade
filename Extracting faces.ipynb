```python
import cv2,os,time
cap = cv2.VideoCapture(0)
Name = input("Enter the candidate name")
try:
    if not os.path.exists(Name):
        os.makedirs(Name)
except OSError:
    print ('Error: Creating directory of data')
currentFrame = 1
FaceFrontalCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
FaceSideCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_profileface.xml")
try:
    while(currentFrame <= 30):
        # Capture frame-by-frame
        ret, frame = cap.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = FaceFrontalCascade.detectMultiScale(gray, scaleFactor=1.31,minNeighbors=10)
        for (x,y,w,h) in faces:
            image = cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
            roi_gray = image[y:y+h, x:x+w]
            print("[INFO] Object found. Saving locally.")
            xd = cv2.resize(roi_gray , (32,32))
            grayscale = cv2.cvtColor(xd, cv2.COLOR_BGR2GRAY)
            name = './'+str(Name)+'/' + str(currentFrame) + '.jpg'
            print ('Creating...' + name)
            cv2.imwrite(name, grayscale)
            currentFrame += 1
        time.sleep(0.5)
    while(currentFrame <= 60):
        # Capture frame-by-frame
        ret, frame = cap.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = FaceSideCascade.detectMultiScale(gray, scaleFactor=1.31,minNeighbors=10)
        for (x,y,w,h) in faces:
            image = cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
            roi_gray = image[y:y+h, x:x+w]
            print("[INFO] Object found. Saving locally.")
            xd = cv2.resize(roi_gray , (32,32))
            grayscale = cv2.cvtColor(xd, cv2.COLOR_BGR2GRAY)
            name = './'+str(Name)+'/' + str(currentFrame) + '.jpg'
            print ('Creating...' + name)
            cv2.imwrite(name, grayscale)
            currentFrame += 1
        time.sleep(0.5)       
#if theres a keyboard interrupt, release cap   
except (KeyboardInterrupt, SystemExit):
    cap.release()
    cv2.destroyAllWindows()
# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
